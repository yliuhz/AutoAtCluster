```python
import pickle as pkl

import networkx as nx
import numpy as np
import scipy.sparse as sp
# import torch
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, average_precision_score
import sklearn.preprocessing as preprocess
import sys
def load_data(dataset_str): # {'pubmed', 'citeseer', 'cora'}
    """Load data."""

    if dataset_str == 'wiki':
        adj, features, label = load_wiki()
        return adj, features, label, 0, 0, 0

    elif dataset_str in ["cora", "citeseer", "pubmed"]:

        names = ['x', 'y', 'tx', 'ty', 'allx', 'ally', 'graph']
        objects = []
        for i in range(len(names)):
            with open("/data/liuyue/New/AGE/data/ind.{}.{}".format(dataset_str, names[i]), 'rb') as f:
                if sys.version_info > (3, 0):
                    objects.append(pkl.load(f, encoding='latin1'))
                else:
                    objects.append(pkl.load(f))

        x, y, tx, ty, allx, ally, graph = tuple(objects)
        test_idx_reorder = parse_index_file("/data/liuyue/New/AGE/data/ind.{}.test.index".format(dataset_str))
        test_idx_range = np.sort(test_idx_reorder)

        if dataset_str == 'citeseer':
            # Fix citeseer dataset (there are some isolated nodes in the graph)
            # Find isolated nodes, add them as zero-vecs into the right position
            test_idx_range_full = range(min(test_idx_reorder), max(test_idx_reorder)+1)
            tx_extended = sp.lil_matrix((len(test_idx_range_full), x.shape[1]))
            tx_extended[test_idx_range-min(test_idx_range), :] = tx
            tx = tx_extended
            ty_extended = np.zeros((len(test_idx_range_full), y.shape[1]))
            ty_extended[test_idx_range-min(test_idx_range), :] = ty
            ty = ty_extended

        features = sp.vstack((allx, tx)).tolil()
        features[test_idx_reorder, :] = features[test_idx_range, :]
        adj = nx.adjacency_matrix(nx.from_dict_of_lists(graph))

        labels = np.vstack((ally, ty))
        labels[test_idx_reorder, :] = labels[test_idx_range, :]

        idx_test = test_idx_range.tolist()
        idx_train = range(len(y))
        idx_val = range(len(y), len(y)+500)


        if labels.ndim > 1:
            if labels.shape[1] == 1:
                labels = labels.view(-1)
            else:
                labels = labels.argmax(1)

    elif dataset_str in ["amazon-photo", "amazon-computers", "cora-full"]:
        map2names = {
            "amazon-photo": "/data/liuyue/New/SBM/mySBM/data/amazon_electronics_photo.npz",
            "amazon-computers": "/data/liuyue/New/SBM/mySBM/data/amazon_electronics_computers.npz",
            "cora-full": "/data/liuyue/New/SBM/mySBM/data/cora_full.npz",
        }

        data = np.load(map2names[dataset_str])
        # print(list(data.keys()))
        adj_data, adj_indices, adj_indptr, adj_shape = data["adj_data"], data["adj_indices"], data["adj_indptr"], data["adj_shape"]
        attr_data, attr_indices, attr_indptr, attr_shape = data["attr_data"], data["attr_indices"], data["attr_indptr"], data["attr_shape"]
        labels = data["labels"]

        adj = sp.csr_matrix((adj_data, adj_indices, adj_indptr), shape=adj_shape).tocoo()
        features = sp.csr_matrix((attr_data, attr_indices, attr_indptr), shape=attr_shape).tocoo()

        if labels.ndim > 1:
            if labels.shape[1] == 1:
                labels = labels.reshape(-1)
            else:
                labels = labels.argmax(1)

        return adj, features, labels, 0, 0, 0

    return adj, features, labels, idx_train, idx_val, idx_test

def load_wiki():
    f = open('./AGE/data/graph.txt','r')
    adj, xind, yind = [], [], []
    for line in f.readlines():
        line = line.split()
        
        xind.append(int(line[0]))
        yind.append(int(line[1]))
        adj.append([int(line[0]), int(line[1])])
    f.close()
    ##print(len(adj))

    f = open('./AGE/data/group.txt','r')
    label = []
    for line in f.readlines():
        line = line.split()
        label.append(int(line[1]))
    f.close()

    f = open('./AGE/data/tfidf.txt','r')
    fea_idx = []
    fea = []
    adj = np.array(adj)
    adj = np.vstack((adj, adj[:,[1,0]]))
    adj = np.unique(adj, axis=0)
    
    labelset = np.unique(label)
    labeldict = dict(zip(labelset, range(len(labelset))))
    label = np.array([labeldict[x] for x in label])
    adj = sp.csr_matrix((np.ones(len(adj)), (adj[:,0], adj[:,1])), shape=(len(label), len(label)))

    for line in f.readlines():
        line = line.split()
        fea_idx.append([int(line[0]), int(line[1])])
        fea.append(float(line[2]))
    f.close()

    fea_idx = np.array(fea_idx)
    features = sp.csr_matrix((fea, (fea_idx[:,0], fea_idx[:,1])), shape=(len(label), 4973)).toarray()
    scaler = preprocess.MinMaxScaler()
    #features = preprocess.normalize(features, norm='l2')
    features = scaler.fit_transform(features)
#     features = torch.FloatTensor(features)

    return adj, features, label
def parse_index_file(filename):
    index = []
    for line in open(filename):
        index.append(int(line.strip()))
    return index
def sample_mask(idx, l):
    """Create mask."""
    mask = np.zeros(l)
    mask[idx] = 1
    return np.array(mask, dtype=np.bool)
```

```python
true_labels = {}
num_edges = {}
for dataset in datasets:
    adj, _, label, _, _, _ = load_data(dataset)
    true_labels[dataset] = label
    num_edges[dataset] = adj.sum()
    print("{}: {} {}".format(dataset, label.shape, num_edges[dataset]))
```

```python
import numpy as np
from sklearn.metrics import normalized_mutual_info_score as NMI, adjusted_mutual_info_score as AMI, adjusted_rand_score as ARI
from sklearn.cluster import KMeans, SpectralClustering
from sklearn.preprocessing import normalize, MinMaxScaler
from tqdm import tqdm
from sklearn.manifold import TSNE
import random
import os

os.makedirs("Cluster", exist_ok=True)


for model_idx, model in tqdm(enumerate(models), total=len(models)):
    os.makedirs("Cluster/{}".format(model), exist_ok=True)
    
    for dataset in datasets:
        nclass = nclasses[dataset]
#         cluster = KMeans(n_clusters=nclass)
        
        if model == "GRACE":
            dataset_ = GRACE_datasets[dataset]
        else:
            dataset_ = dataset
        
        nmis, amis, aris = [], [], []
        for seed in range(10):
            np.random.seed(seed)
            random.seed(seed)
            
            print(model, dataset, seed)
            
            # -- KMeans -- #
            cluster_name = "SC"
#             cluster = KMeans(n_clusters=nclass)
            cluster = SpectralClustering(n_clusters=nclass, affinity="precomputed", random_state=seed)
            if not os.path.exists("Cluster/{}/{}_{}_preds_{}.npz".format(model, cluster_name, dataset, seed)):
                emb_path = emb_paths[model]
                emb_path = os.path.join(emb_path, "{}_{}_emb_{}.npz".format(model, dataset_, seed))
            
                try:
                    data = np.load(emb_path)
                    emb = data["emb"]
                    labels = true_labels[dataset]
                    
                    if cluster_name == "SC":
                        if model != "AGE":
                            zmax = emb.max(axis=1, keepdims=True)
                            zmin = emb.min(axis=1, keepdims=True)
                            diff = zmax - zmin
                            diff[diff == 0] = 1.
                            f_adj = (emb - zmin) / diff
                            f_adj = normalize(f_adj)
                        else:
                            f_adj = emb
                        f_adj = np.matmul(f_adj, np.transpose(f_adj))
                    else:
                        f_adj = emb

                    preds = cluster.fit_predict(f_adj)
                    np.savez("Cluster/{}/{}_{}_preds_{}.npz".format(model, cluster_name, dataset, seed), preds=preds, labels=labels)
                except Exception as e:
                    print(e.__class__, e)
                    continue
            else:
                data = np.load("Cluster/{}/{}_{}_preds_{}.npz".format(model, cluster_name, dataset, seed))
                preds, labels = data["preds"], data["labels"]
            
                
            nmi = NMI(labels, preds)
            ami = AMI(labels, preds)
            ari = ARI(labels, preds)
                
            nmis.append(nmi)
            amis.append(ami)
            aris.append(ari)
            
        mean_nmi = np.mean(nmis)
        mean_ami = np.mean(amis)
        mean_ari = np.mean(aris)
        
        df_nmi["{}_{}".format(dataset, cluster_name)][model_idx] = mean_nmi
        df_ami["{}_{}".format(dataset, cluster_name)][model_idx] = mean_ami
        df_ari["{}_{}".format(dataset, cluster_name)][model_idx] = mean_ari

                
```

```python
import numpy as np
from sklearn.preprocessing import normalize, MinMaxScaler
# import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.cluster import SpectralClustering
from sklearn.metrics import normalized_mutual_info_score as NMI, adjusted_mutual_info_score as AMI, adjusted_rand_score as ARI
# import matplotlib.pyplot as plt
# import scienceplots
# plt.style.use('science')



plot_models = ["GRACE"]
plot_datasets = ["cora", "citeseer", "wiki", "pubmed"]
# plot_datasets = ["cora", "citeseer", "wiki"]

# fig, axs = plt.subplots(len(plot_models), len(plot_datasets), figsize=(14,2))


def build_bars(sim, labels):
    Y_sim = np.zeros((labels.shape[0], labels.shape[0]), dtype=int)
    for i in range(labels.shape[0]):
        Y_sim[i] = (labels[i] == labels)
    Y_sim = Y_sim.reshape(-1)
    
    sim_neg = ((sim < 0)).sum()
    sim_0_01 = ((sim >=0) * (sim < 0.1)).sum()
    sim_01_05 = ((sim >= 0.1) * (sim < 0.5)).sum()
    sim_05_09 = ((sim >= 0.5) * (sim < 0.9)).sum()
    sim_09_1 = ((sim >= 0.9)).sum()
    
    sim_neg_T = ((sim < 0) * (Y_sim == 1)).sum()
    sim_0_01_T = ((sim >=0) * (sim < 0.1) * (Y_sim == 1)).sum()
    sim_01_05_T = ((sim >= 0.1) * (sim < 0.5) * (Y_sim == 1)).sum()
    sim_05_09_T = ((sim >= 0.5) * (sim < 0.9) * (Y_sim == 1)).sum()
    sim_09_1_T = ((sim >= 0.9) * (Y_sim == 1)).sum()
    
    sim_neg_N = ((sim < 0) * (Y_sim == 0)).sum()
    sim_0_01_N = ((sim >=0) * (sim < 0.1) * (Y_sim == 0)).sum()
    sim_01_05_N = ((sim >= 0.1) * (sim < 0.5) * (Y_sim == 0)).sum()
    sim_05_09_N = ((sim >= 0.5) * (sim < 0.9) * (Y_sim == 0)).sum()
    sim_09_1_N = ((sim >= 0.9) * (Y_sim == 0)).sum()
    
    height = np.array([sim_neg, sim_0_01, sim_01_05, sim_05_09, sim_09_1]) / sim.shape[0]
    height_T = np.array([sim_neg_T, sim_0_01_T, sim_01_05_T, sim_05_09_T, sim_09_1_T]) / sim.shape[0]
    height_N = np.array([sim_neg_N, sim_0_01_N, sim_01_05_N, sim_05_09_N, sim_09_1_N]) / sim.shape[0]
    
    index = np.arange(5) + 0.3
    index_T = np.arange(5) + 0.3 - 0.125
    index_N = np.arange(5) + 0.3 + 0.125
    
    yticks = np.arange(0.0, 1.0, 0.2)
    
    return index, index_T, index_N, height, height_T, height_N, yticks

def link_infer(emb, model_name, dataset, seed):
    import torch
    import torch.nn as nn
    import torch.nn.functional as F
#     class BinaryClassification(nn.Module):
#         def __init__(self, in_dim, hidden_dim, dropout=0.1):
#             super(BinaryClassification, self).__init__()
#             # Number of input features is 12.
#             self.layer_1 = nn.Linear(in_dim, hidden_dim) 
#             self.layer_2 = nn.Linear(hidden_dim, hidden_dim)
#             self.layer_out = nn.Linear(hidden_dim, 1) 

#             self.relu = nn.ReLU()
#             self.dropout = nn.Dropout(p=dropout)
#             self.batchnorm1 = nn.BatchNorm1d(hidden_dim)
#             self.batchnorm2 = nn.BatchNorm1d(hidden_dim)

#         #     self.init_weights()

#         # def init_weights(self):
#         #     nn.init.xavier_normal_(self.layer_1.weight, gain=1.414)
#         #     nn.init.xavier_normal_(self.layer_2.weight, gain=1.414)
#         #     nn.init.xavier_normal_(self.layer_out.weight, gain=1.414)

#         #     # nn.init.xavier_normal_(self.batchnorm1.weight, gain=1.414)
#         #     # nn.init.xavier_normal_(self.batchnorm2.weight, gain=1.414)


#         def forward(self, inputs):
#             x = self.relu(self.layer_1(inputs))
#             x = self.batchnorm1(x)
#             x = self.relu(self.layer_2(x))
#             x = self.batchnorm2(x)
#             x = self.dropout(x)
#             x = self.layer_out(x)

#             return x.reshape(-1)
    class BinaryClassification(nn.Module):
        def __init__(self, in_dim, hidden_dim, n_layers=2, dropout=0.2):
            super(BinaryClassification, self).__init__()
            # Number of input features is 12.
            self.layers = nn.ModuleList()
            self.layers.append(nn.Linear(in_dim, hidden_dim))
            for i in range(n_layers-2): 
                self.layers.append(nn.Linear(hidden_dim, hidden_dim))
            self.layer_out = nn.Linear(hidden_dim, 1) 

            self.relu = nn.ReLU()
            self.dropout = nn.Dropout(p=dropout)
            self.batchnorm1 = nn.BatchNorm1d(hidden_dim)
            self.batchnorm2 = nn.BatchNorm1d(hidden_dim)

        #     self.init_weights()

        # def init_weights(self):
        #     nn.init.xavier_normal_(self.layer_1.weight, gain=1.414)
        #     nn.init.xavier_normal_(self.layer_2.weight, gain=1.414)
        #     nn.init.xavier_normal_(self.layer_out.weight, gain=1.414)

        #     # nn.init.xavier_normal_(self.batchnorm1.weight, gain=1.414)
        #     # nn.init.xavier_normal_(self.batchnorm2.weight, gain=1.414)


        def forward(self, inputs):
            x = inputs
            for layer in self.layers:
                x = layer(x)
                x = self.relu(x)
                # x = self.layer_1(inputs)
                # x = self.relu(x)
                # x = self.batchnorm1(x)
                # x = self.layer_2(x)
                # x = self.relu(x)
                # x = self.batchnorm2(x)
            x = self.dropout(x)
            x = self.layer_out(x)

            return x.reshape(-1)
    try:
        model = BinaryClassification(emb.shape[-1], 500, n_layers=3).to("cuda:0")
        model.load_state_dict(torch.load("../saved_models/{}-{}-{}.statedict".format(model_name, dataset, seed)))
    except Exception:
        model = BinaryClassification(emb.shape[-1], 500, n_layers=2).to("cuda:0")
        model.load_state_dict(torch.load("../saved_models/{}-{}-{}.statedict".format(model_name, dataset, seed)))
    model.eval()
    inx = torch.FloatTensor(emb).to("cuda:0")
    
    with torch.no_grad():
        sim = np.zeros((emb.shape[0], emb.shape[0]))
        for i in tqdm(range(emb.shape[0]), total=emb.shape[0]):
            inputs = (inx[i] * inx).to("cuda:0")

            # zmax = inputs.max(dim=1, keepdim=True)[0]
            # zmin = inputs.min(dim=1, keepdim=True)[0]
            # inputs = (inputs - zmin) / (zmax - zmin)
#             inputs = F.normalize(inputs)

            preds = torch.sigmoid(model(inputs)).cpu().numpy()
            
            sim[i] = preds

    return sim

for model_idx, model in enumerate(plot_models):
    for dataset_idx, dataset in enumerate(plot_datasets):
        emb_heights = np.zeros(5)
        lk_heights = np.zeros(5)
        
        nseed = 1
        for seed in range(nseed):
            print(model, dataset, seed)
            
            if model == "GRACE":
                dataset_ = GRACE_datasets[dataset]
            else:
                dataset_ = dataset

            emb_path = emb_paths[model]
            emb_path = os.path.join(emb_path, "{}_{}_emb_{}.npz".format(model, dataset_, seed))
            data = np.load(emb_path)
            emb = data["emb"]
            if model != "AGE":
                sim = normalize(emb)
            else:
                sim = emb
            sim = np.matmul(sim, np.transpose(sim)).reshape(-1)
            labels = true_labels[dataset]
            
#             cluster = SpectralClustering(n_clusters=nclasses[dataset], affinity="precomputed", random_state=seed)
#             preds = cluster.fit_predict(sim.reshape(labels.shape[0], -1))
#             nmi = NMI(labels, preds)
#             ami = AMI(labels, preds)
#             ari = ARI(labels, preds)
#             print(nmi, ami, ari)


            index, index_T, index_N, height, height_T, height_N, yticks = build_bars(sim, labels)
            emb_heights += height / nseed
            
            sim = link_infer(emb, model, dataset, seed).reshape(-1)
            index, index_T, index_N, height, height_T, height_N, yticks = build_bars(sim, labels)
            lk_heights += height / nseed
        
        fig = plt.figure(figsize=(3.5,2))
        ax = plt.gca()
#         if len(plot_datasets) > 1:
#             ax = axs[dataset_idx]
#         else:
#             ax = axs
        ax.bar(index-0.125, emb_heights, color="#E3918F", width=0.25, linewidth=0.5, edgecolor="black", label="$S_{uv}$")
#         ax.bar(index_T, height_T, color="#E3918F", width=0.25, linewidth=0.5, edgecolor="black", label="True")
#         ax.bar(index_N, height_N, color="#B2B2FA", width=0.25, linewidth=0.5, edgecolor="black", label="False")
        ax.bar(index+0.125, lk_heights, color="#B2B2FA", width=0.25, linewidth=0.5, edgecolor="black", label="$O_{uv}$")
        ax.set_xticks(index)
#         ax.set_yticks(yticks)
#         ax.set_xticklabels(["", "", "", "", ""], rotation=0)
#         ax.set_yticklabels(["", "", "", "", ""], rotation=0)
        ax.set_xticklabels(["<0", "0-0.1", "0.1-0.5", "0.5-0.9", ">0.9"], rotation=25)
        ax.set_xlabel("Similarity interval", fontsize=12)
        ax.set_ylabel("Frequency", fontsize=12)
#         if dataset_idx == 0:
#             ax.set_ylabel("Frequency")
#         if model_idx == 0:
#             ax.title.set_text("{}".format(dataset))
        ax.legend(loc="upper right")
        
#         s_min = sim.min()
#         s_max = sim.max()
#         sim = (sim - s_min) / (s_max - s_min)
#         index, height = build_bars(sim)
#         ax = axs[1][dataset_idx]
#         ax.bar(index, height, color="#B2B2F9", width=0.25)
#         ax.set_xticks(index)
#         ax.set_xticklabels(["<0", "0-0.1", "0.1-0.5", "0.5-0.9", ">0.9"], rotation=0)
#         ax.set_xlabel("Similarity Interval")
#         if dataset_idx == 0:
#             ax.set_ylabel("Frequency")
#         ax.tick_params(
#             axis='both',          # changes apply to the x-axis
#             which='both',      # both major and minor ticks are affected
#             bottom=False,      # ticks along the bottom edge are off
#             top=False,         # ticks along the top edge are off
#             left=False,
#             labelbottom=False) # labels along the bottom edge are off
#         ax.spines.right.set_visible(False)
#         ax.spines.top.set_visible(False)
#         ax.spines.left.set_visible(False)
#         ax.set_xlim(index[1]-0.2, index[-1]+0.2)
        plt.savefig("Paper/chart1_{}_{}_output.pdf".format(model, dataset), bbox_inches='tight')
        plt.show()
        plt.clf()

# import os
# os.makedirs("Paper", exist_ok=True)
# plt.savefig("Paper/chart1_{}.pdf".format(model), bbox_inches='tight')
# plt.show()
```

```python
from tqdm import tqdm
import numpy as np
def sampling(adj, rate=0.5):
    n = adj.shape[0]
    adj = adj.toarray()
    
    ret = np.zeros((n,n))
    
    for i in range(n):
        row_idx = adj[i].nonzero()[0]
        arr = np.random.choice(row_idx, int(rate*row_idx.shape[0]))
        ret[i][arr] = 1
    
    return sp.coo_matrix(ret)
```

```python
def louvain_cluster(adj, labels, random_state=None):
    from community import community_louvain
    import networkx as nx
    from sklearn.metrics import normalized_mutual_info_score as NMI, adjusted_mutual_info_score as AMI

    graph = nx.from_scipy_sparse_matrix(adj)
    partition = community_louvain.best_partition(graph, random_state=random_state)
    preds = list(partition.values())

    return preds
```

```python
import numpy as np
import random
import scipy.sparse as sp

from sklearn.metrics import normalized_mutual_info_score as NMI, adjusted_mutual_info_score as AMI, adjusted_rand_score as ARI

xx = np.arange(1, 11, dtype=int).tolist()
xx = [str(x) + "m" for x in xx]
df_data_nmi = pd.DataFrame(columns=["models", "dataset"]+xx)
df_data_ami = pd.DataFrame(columns=["models", "dataset"]+xx)
df_data_ari = pd.DataFrame(columns=["models", "dataset"]+xx)

for model_idx, model in enumerate(plot_models):
    for dataset in plot_datasets:
        nseed=10
        labels = true_labels[dataset]
        
        nmi_m, ami_m, ari_m, nclass_m = {}, {}, {}, {}
        
        edges = np.arange(num_edges[dataset], 11*num_edges[dataset], num_edges[dataset], dtype=int)
        for m_idx, m in enumerate(edges):
            nmis, amis, aris, nclasses = [], [], [], []

            for seed in range(nseed):
                np.random.seed(seed)
                random.seed(seed)

#                 print(model, dataset, m_idx, seed)
                
                data = np.load("{}/Cluster_/{}/lo_{}_preds_{}_{}.npz".format(model, model, dataset, seed, m))
                preds = data["preds"]
                
                nmi = NMI(labels, preds)
                ami = AMI(labels, preds)
                ari = ARI(labels, preds)
                nclass = np.unique(preds).shape[0]
                
                nmis.append(nmi)
                amis.append(ami)
                aris.append(ari)
                nclasses.append(nclass)
            
            if len(nmis) > 0:
                nmi_m[m] = np.mean(nmis)
                ami_m[m] = np.mean(amis)
                ari_m[m] = np.mean(aris)
                nclass_m[m] = np.mean(nclasses)
                
        new_line = pd.DataFrame([[model, dataset] + list(nmi_m.values())], columns=df_data_nmi.columns)
        df_data_nmi = pd.concat([df_data_nmi, new_line])
        
        new_line = pd.DataFrame([[model, dataset] + list(ami_m.values())], columns=df_data_ami.columns)
        df_data_ami = pd.concat([df_data_ami, new_line])
        
        new_line = pd.DataFrame([[model, dataset] + list(ari_m.values())], columns=df_data_ari.columns)
        df_data_ari = pd.concat([df_data_ari, new_line])
        print(nclass_m.values())

        
df_data_nmi.to_csv("AGE_baseline_NMI.csv", index=False)
df_data_ami.to_csv("AGE_baseline_AMI.csv", index=False)
df_data_ari.to_csv("AGE_baseline_ARI.csv", index=False)
```

```python
from vis import plot_superadj
def load_cora_full_im(rate=0.1, seed=None):
    filename = "/data/liuyue/New/SBM/mySBM/data_im/cora-full_{:.1f}_{:d}.npz".format(rate, seed) # seed=2
    data = np.load(filename)

    adj_raw, features_raw, labels_raw, _, _, _ = load_data("cora-full")

    adj_data, adj_row, adj_col, features_load, labels_load, mask = data["data"], data["row"], data["col"], data["features"], data["labels"], data["mask"]
    adj_load = sp.coo_matrix((adj_data, (adj_row, adj_col)), shape=(labels_load.shape[0], labels_load.shape[0]))

    adj_mask = adj_raw.toarray()[mask,:][:,mask]
    assert (adj_mask - adj_load).sum() < 1e-7
    features_mask = features_raw.toarray()[mask]
    assert (features_mask - features_load).sum() < 1e-7

    return adj_load, features_load, labels_load, mask
```
