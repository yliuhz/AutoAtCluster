```python
import numpy as np
import pandas as pd
```

```python
models = [
    "GAE",
    "VGAE",
    "ARGA",
    "ARVGA",
    "AGE",
    "DGI",
    "MVGRL",
    "GRACE",
    "GGD"
]
```

```python
datasets = [
    "cora",
    "citeseer",
    "wiki",
    "pubmed",
    "amazon-photo",
    "amazon-computers"
]
```

```python
df = pd.DataFrame(data={"models":models}, columns=["models", "cora_pos", "cora_neg", "citeseer_pos", "citeseer_neg", "wiki_pos", "wiki_neg", "pubmed_pos", "pubmed_neg", "amazon-photo_pos", "amazon-photo_neg", "amazon-computers_pos", "amazon-computers_neg"])
df
```

```python
for model_idx, model in enumerate(models):
    for dataset in datasets:
        print(model, dataset)
        
        try:
            data = np.load("{}-{}.npz".format(model, dataset))
            pos_r, neg_r = data["pos"], data["neg"]

            df["{}_pos".format(dataset)][model_idx] = pos_r
            df["{}_neg".format(dataset)][model_idx] = neg_r
        except Exception as e:
            print(e)
```
